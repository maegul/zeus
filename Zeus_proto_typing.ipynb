WHy do these two not equate?  Perhaps because in the first case, above, I am finding the average of two sds, whilst in the second case I am finding the total sd for the combined data set, both of which being separate values and calculations (? ... see [Forum Chat](http://www.talkstats.com/showthread.php/14523-An-average-of-standard-deviations))

```python
def sum_var(nx, ny, sx, sy, mx, my):
    num = nx**2*sx**2 + ny**2*sy**2 - ny*sx**2 - ny*sy**2 - nx*sx**2 - nx*sy**2 + ny*nx*sx**2 + ny*nx*sy**2 + nx*ny*(mx-my)**2
    den = (nx+ny) * (nx+ny)
    return ( num / float(den)) ** 0.5
```

They are the same!!  Seems to rely on n-1 degrees of freedom.  To match, the input sd must be ddof=1 and the manual calculation must be ddof=1. THis is with the denominator being $(n_{x}+n_{y}-1)(n_{x}+n_{y})$ 

Else, alter the denominator to $(n_{x}+n_{y})^2$ or $(n_{x}+n_{y})(n_{x}+n_{y})$, and input ddof=1 sd values, the return will match ddof=0 manual calculation.

```python
def sum_std(x, y, ddof=0):
    nx = x.size
    ny = y.size
    sx = x.std(ddof=1)
    sy = y.std(ddof=1)
    mx = x.mean()
    my = y.mean()
    
    num = nx**2*sx**2 + ny**2*sy**2 - ny*sx**2 - ny*sy**2 - nx*sx**2 - nx*sy**2 + ny*nx*sx**2 + ny*nx*sy**2 + nx*ny*(mx-my)**2
    den = (nx+ny-ddof) * (nx+ny)
    return ( num / float(den)) ** 0.5
```

```python
def sum_std_stat(nx, ny, sx, sy, mx, my, ddof=0):
    num = nx**2*sx**2 + ny**2*sy**2 - ny*sx**2 - ny*sy**2 - nx*sx**2 - nx*sy**2 + ny*nx*sx**2 + ny*nx*sy**2 + nx*ny*(mx-my)**2
    den = (nx+ny-ddof) * (nx+ny)
    return ( num / float(den)) ** 0.5
```

##Sum of Variance for Uncorrelated Variables

So we have the sum of variances:

$\sigma_{X+Y}^{2} = \sigma_{X}^{2} + \sigma_{Y}^{2}$

Which is true where $X$ and $Y$ are **uncorrelated or independent** random variables.

More generally, this can be expressed as

$Var(\sum\limits_{i=1}^{N} X_{i}) = \sum\limits_{i=1}^{N} Var(X_{i})$

Where $X_{1} \ldots X_{N}$ are the independent variables.

---

##Tangent

Curiously, applying this to the mean of a variable leads to the central limit theorem as dividing by $N$ is linear ...

$Var(\bar{X}) = Var(\frac{1}{N}\sum\limits_{i=1}^{N} X_{i}) = \frac{1}{N^2}\sum\limits_{i=1}^{N} Var(X_i)$

if the all the variables have the same variance, which they should if they are from the same population, then ...

$\frac{1}{N^2}\sum\limits_{i=1}^{N} Var(X_i)= \frac{1}{N^2} N\sigma^2 = \frac{\sigma^2}{N}$, the square root of which is the *standard error of the mean*.

---

##Correlated Variables

*However*, if the variables are **not** independent but are **correlated**, then the variance of their sum involves the covariance, which is a statistic that quantifies the extent to which variables vary *together*.

For two variables, 

$\sigma_{X+Y}^{2} = \sigma_{X}^{2} + \sigma_{Y}^{2} + \sigma_{X, Y}^2$

where $ \sigma_{X, Y}^2 $ is the covariance between $X$ and $Y$.

Regarding the specific definitions of variance and covariance, note that

$Var(X) = \frac{1}{N}\sum\limits_{i=1}^N(X_i - \bar{X})^2$.

(*More specifically, this is the 'biased' sample variance, where to unbias, the degrees of freedom would have to be reduced by one, but otherwise, in as much as this equates to the mean square of differences from the mean, it defines variance; the same holds for covariance below*)

$Cov(X, Y) = \frac{1}{N}\sum\limits_{i=1}^N(X_i - \bar{X})(Y_i - \bar{Y})$

Curiously, note that given the above, $Var(X) = Cov(X, X)$; and that Pearson's correlation is a normalised covariance given by,

$p_{X, Y} = \frac{Cov(X, Y)}{\sigma_X\sigma_Y} = \frac{\frac{1}{N}\sum\limits_{i=1}^N(X_i - \bar{X})(Y_i - \bar{Y})}{\sqrt{\frac{1}{N}\sum\limits_{i=1}^N(X_i - \bar{X})^2}\sqrt{\frac{1}{N}\sum\limits_{i=1}^N(Y_i - \bar{Y})^2}}$


Getting back to the problem, the variance of the sum of *multiple* correlated variables is given by,

$Var(\sum\limits_{i=1}^{N} X_{i}) = \sum\limits_{i=1}^{N}Var(X_i) + 2\sum\limits_{i=1}^N\sum\limits_{j=i+1}^{N}Cov(X_i, X_j)$

where weighting the variables alters the equation only linearly:

$Var(\sum\limits_{i=1}^{N} a_{i}X_{i}) = \sum\limits_{i=1}^{N}a_i^2Var(X_i) + 2\sum\limits_{i=1}^N\sum\limits_{j=i+1}^{N}a_ia_jCov(X_i, X_j)$

---

Bottom line is that I can either add bins together and calculate the variance as though they are uncorrelated, which is likely to be false, or calculate the variance as though they are correlated, which is probably best.  

If I am employing a kernel, such as a gaussian, then the weightings I employ can also be reflected using the above.  This would require me to do the gaussian filtering manually.

```python
fig, ax = plt.subplots()
fig.subplots_adjust(bottom=0.2, left=0.1)
ax.set_xlim(-2*np.pi, 2*np.pi)
ax.set_ylim(-5, 5)

ax.xaxis.set_ticks([-2*np.pi, -1*np.pi, 0, np.pi, 2*np.pi])

ax.xaxis.set_ticklabels(['$-2\pi$', '$-\pi$', '$0$', '$\pi$', '$2\pi$'], fontsize=16)

ax.xaxis.grid(True)
ax.yaxis.grid(True)

freq_i = 1
amp_i = 1

freq = freq_i
amp = amp_i

def func(x, freq, amp):
    return amp * np.sin(freq * x)

t = np.linspace(-10, 10, 1000)
line, = plt.plot(t, func(t, freq_i, amp_i), lw=2)

def on_change_a(val):
    global amp
    amp = val
    line.set_ydata(func(t, freq, amp))
    ax.set_title('$f(x) = %s * \sin( %s * x) $' %(np.around(amp, 2), np.around(freq, 2)))

def on_change_f(val):
    global freq
    freq = val
    line.set_ydata(func(t, freq, amp))
    ax.set_title('$f(x) = %s * \sin( %s * x) $' %(np.around(amp, 2), np.around(freq, 2)))    
    
    
slider_ax_a = plt.axes([0.1, 0.05, 0.8, 0.02])
slider_a = Slider(slider_ax_a, "amplitude", 0, 10, valinit=1, color='#AAAAAA')
slider_a.on_changed(on_change_a)

slider_ax_f = plt.axes([0.1, 0.1, 0.8, 0.02])
slider_f = Slider(slider_ax_f, "frequency", 0, 10, valinit=1, color='#AAAAAA')
slider_f.on_changed(on_change_f)
```

```python
fig, ax = plt.subplots(1,2)
fig.subplots_adjust(bottom=0.2, left=0.1, hspace=0.7)
ax[0].set_xlim(-5, 5)
ax[0].set_ylim(-5, 5)
ax[1].set_xlim(-5, 5)
ax[1].set_ylim(-5, 5)
ax[0].xaxis.grid(True)
ax[0].yaxis.grid(True)
ax[1].xaxis.grid(True)
ax[1].yaxis.grid(True)

val_i = 1

def func(x, val):
    return val*x*(x**2 - 3)
def func_d(x, val):
    return 3*val*(x**2) - val*3

t = np.linspace(-10, 10, 1000)
line_func, = ax[0].plot(t, func(t, val_i), lw=2)
line_func_d, = ax[1].plot(t, func_d(t, val_i), lw=2)


def on_change(val):
    line_func.set_ydata(func(t, val))
    line_func_d.set_ydata(func_d(t, val))
    ax[0].set_title('$f(x) = %s * x * (x^2 - 3)$' %(np.around(val, 2)))
    ax[1].set_title(r'$\frac{dy}{dx} = %s * 3 * x^2 - %s * 3$' %(np.around(val, 2), np.around(val,2)), fontsize=14)

    
    
slider_ax = plt.axes([0.1, 0.1, 0.8, 0.02])
slider = Slider(slider_ax, "Offset", -1.5, 1.5, valinit=1, color='#AAAAAA')
slider.on_changed(on_change)
```
